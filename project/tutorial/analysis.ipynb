```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('science')
```

### Rabi

```python
rabi = np.loadtxt('rabi.txt')
rabi = [ls[1:] for ls in rabi]
```

```python
plt.plot(rabi[0], rabi[1])
plt.plot(rabi[0], rabi[2])
```

```python
rabi_contrast = [sig/ref for sig,ref in zip(rabi[1], rabi[2])]
plt.figure(figsize=(10,7))
plt.plot(rabi[0], rabi_contrast, 'o--', markersize=3)
plt.title('Rabi oscillation')
plt.xlabel('MW duation (ns)')
plt.ylabel('sig / ref')
```
